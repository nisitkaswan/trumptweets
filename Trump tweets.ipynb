```python
import pandas as pd
import json
```

```python
df1=pd.read_json('https://query.data.world/s/kpf4LLqAll5dYbMpV5WYbXkwQh6Bpq')

df2=pd.read_json('https://query.data.world/s/dYtdFJPEQdsAbA8oXDYTQJj0cWWoAR')

df3=pd.read_json('https://query.data.world/s/43i9j7GxCK7wCl1InvdLAgLsvEuewx')

#merging the dataset
df4=[df1, df2, df3]
df=pd.concat(df4)
```

```python
#finding all references to Donald Trump
ref=df[df['text'].str.contains("@realDonaldTrump")]
#printing all references
print("The account referencing donald trump are:\n")
print(ref)
```

```python
#total number of accounts
total_accounts=df.groupby('user_id').size().shape
#total number of accounts referencing donald trump
donaldtrump_related_accounts=ref.groupby('user_id').size().shape
```

```python
#total percentage of accounts tweeting about Donald Trump
total_percentage=(float(donaldtrump_related_accounts[0])/float(total_accounts[0]))*100
#p total percentage
test=str(total_percentage)
print("% of accounts tweeting about donald trump are "+test)
```

```python
#number of tweets referencing Donald Trump from each user accounts
frequency=ref.groupby(['user_id','screen_name'])['text'].count()
#sorting user'id based on the count 
accounts_in_descending_order=frequency.nlargest(donaldtrump_related_accounts[0])
#printng the accounts
print("The accounts in descending order of frequency:")
print(accounts_in_descending_order)
```
